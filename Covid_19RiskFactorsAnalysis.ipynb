```python
# Create functions that help to clean the data
# Convert upper cases to lower cases
def lower(data):
    data = data.lower()
    return data

# Replace symbols, comma by empty spaces
def remove_punctuation(data):
    symbols = "!\"#$%&()*+./:;<=>?@[\]^_`{|}~\n"
    for i in symbols:
        data = data.replace(i, " ")
    #replace the comma seperately
    data = data.replace(',', ' ')
    return data
# Replace apostrophe by empty spaces
def remove_apostrophe(data):
    data = data.replace("'", "")
    return data
# Replace number by empty spaces
def remove_numbers(data):
    nums = "0123456789"
    for i in nums:
        data = data.replace(i, "")
        return data
# Remove stopwords using words() method, and generate a cleaning string of words
def remove_stop_words(data):
    stop_words = stopwords.words('english')
    words = word_tokenize(str(data))
    new_text = ""
    for w in words:
        if w not in stop_words and len(w) > 1:
            new_text = new_text + " " + w
    return new_text
```
